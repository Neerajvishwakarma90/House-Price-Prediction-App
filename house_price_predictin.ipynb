```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('clean_house_data.csv', index_col=False)
```

```python
data.drop(columns=['Unnamed: 0'],inplace=True)
data
```

```python
x = data.drop(columns='price')
y = data['price']
```

```python
display(x,y)
```

```python
import pandas as pd
import numpy as np
import joblib as jb
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
# data = pd.read_csv('clean_house_data.csv')
# data = data.rename(columns={'total_sqft': 'total_sqft'})
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('location', OneHotEncoder(handle_unknown='ignore', sparse_output=False), ['location']),
        ('num', StandardScaler(), ['total_sqft', 'bath', 'balcony', 'bedroom'])
    ],
    remainder='drop'
)

pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])
```

```python
pipeline.fit(x_train, y_train)
print(f"✅ Model trained! Test R²: {pipeline.score(x_test, y_test):.3f}")
```

```python
jb.dump(pipeline, 'model.pkl')
```
