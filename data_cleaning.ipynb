```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Bengaluru_House_Data.csv')
```

```python
data
```

```python
data.drop(columns=['area_type','availability','society',],inplace=True)
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
data.head()
```

```python
data.head()
```

```python
data['location'] = data['location'].apply(lambda x : x.strip())
```

```python
data['location'].value_counts()
```

```python
location_status = data.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_status
```

```python
loc_less_10 = location_status[location_status<=10]
loc_less_10
```

```python
data['location'] = data['location'].apply(lambda x : 'other' if x in loc_less_10 else x)
```

```python
data['location'].value_counts()
```

```python
data
```

```python
data['size'].unique()
```

```python
data['bedrooms'] = data['size'].apply(lambda x:int(x.split(' ')[0]))
```

```python
data
```

```python
data['total_sqft'].unique()
```

```python
def clean(sqft):
  tokens = sqft.split('-')
  if len(tokens) == 2:
    return (float(tokens[0])+float(tokens[1]))/2
  else:
    try:
      return float(sqft)
    except:
      return None
```

```python
data['total_sqft'] = data['total_sqft'].apply(clean)
```

```python
data['total_sqft'].unique()
```

```python
data
```

```python
data.describe()
```

```python
data['sqft_per_bed'] = data['total_sqft']/data['bedrooms'] 
```

```python
data['sqft_per_bed'].describe()
```

```python
data2 = data[data['total_sqft'] >= 300]
```

```python
data2
```

```python
data2['price_per_sqft'] = round(data2['price']*100000/data2['total_sqft'],2)
```

```python
data2
```

```python
data2['price_per_sqft'].describe()
```

```python
data3 = data2[data2['price_per_sqft']>=2000]
```

```python
data3
```

```python
data3.drop(columns=['size','sqft_per_bed','price_per_sqft'],axis=1,inplace=True)
```

```python
data3.to_csv("Benagalure_clean_data.csv")
```
